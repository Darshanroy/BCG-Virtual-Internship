# HYPOTHESIS 

To verify the hypothesis that price sensitivity is correlated with churn, we can define price sensitivity as the ratio of the difference between the forecasted energy prices (peak and off-peak) to the forecasted energy consumption. Price-sensitive customers might be more likely to churn if there are significant fluctuations in energy prices relative to their consumption patterns. Here's how you can calculate and model this price sensitivity:

### Steps to Calculate Price Sensitivity:

1. **Calculate Price Sensitivity for Each Customer:**
   - Calculate the difference between `forecast_price_energy_peak` and `forecast_price_energy_off_peak` for each customer.
   - Divide this price difference by `forecast_cons_12m` to get the price sensitivity ratio.



```python
# Calculate price sensitivity ratio
client_data['price_sensitivity'] = (client_data['forecast_price_energy_peak'] - client_data['forecast_price_energy_off_peak']) / client_data['forecast_cons_12m']
```

### 2. Analyze Price Sensitivity for Churning and Non-Churning Customers:

* Compare the average price sensitivity for customers who churned (churn == 1) and those who did not churn (churn == 0).
* Visualize the price sensitivity distribution for both groups (churning and non-churning customers) using histograms or box plots to observe differences.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compare price sensitivity for churning and non-churning customers
sns.boxplot(x='churn', y='price_sensitivity', data=client_data)
plt.xlabel('Churn')
plt.ylabel('Price Sensitivity')
plt.title('Price Sensitivity for Churning and Non-Churning Customers')
plt.show()
```

# Some possible insights from the scatter plot are:

- There is a **negative correlation** between price sensitivity and customer churn. This means that customers who are more sensitive to price changes are more likely to stop using the service or product.
- The majority of the data points are clustered in the **bottom-left quadrant**, which indicates that most customers have low price sensitivity and low churn rate. This suggests that the service or product is affordable and satisfying for most customers.
- There are some outliers in the **top-left quadrant**, which represent customers who have low price sensitivity but high churn rate. This could imply that these customers have other reasons for leaving the service or product, such as quality, availability, or preference issues.
- There are few data points in the **bottom-right quadrant**, which represent customers who have high price sensitivity but low churn rate. This could indicate that these customers are loyal or satisfied with the service or product despite the price changes, or that they have no other alternatives.

```python
from scipy.stats import ttest_ind
churning_customers=client_data[client_data['churn'] == 1]['price_sensitivity']


non_churning_customers = client_data[client_data['churn'] == 0]['price_sensitivity']

# Remove NaN and Inf values from churning_customers and non_churning_customers
churning_customers = churning_customers.replace([np.inf, -np.inf, np.nan], np.nan).dropna()
non_churning_customers = non_churning_customers.replace([np.inf, -np.inf, np.nan], np.nan).dropna()

t_stat, p_value = ttest_ind(churning_customers, non_churning_customers)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

# Null Hypo
The null hypothesis in this casethat there is no difference in the mean price sensitivity between customers who churned and customers who did not churn. This would imply that price sensitivity is not a factor that affects customer churn. To test this hypothesis, we could use a t-test to compare the means of the two groups and calculate the t-statistic and the p-value. The t-statistic measures how far the observed difference is from the expected difference under the null hypothesis. The p-value measures the probability of obtaining a result at least as extreme as the observed one if the null hypothesis is true. A low p-value (usually less than 0.05) indicates that the null hypothesis is unlikely and should be rejected. A high p-value (usually more than 0.05) indicates that the null hypothesis is plausible and cannot be rejected.

# `Report:`

The t-statistic of 0.6606578752540639 and the p-value of 0.5088423880366462 indicate that there is **no significant difference** between the means of the two groups that were compared in the t-test. This means that we **cannot reject** the null hypothesis that the two groups have the same mean. In other words, the difference between the two groups is likely due to **random chance** rather than a real effect.
